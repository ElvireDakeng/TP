```python
model_var= ['A','B','C','D','E','F','Target']
def load_data(model_var):
    """
    load data set from externanl location
    input: model_var
    output:pandas dataframe
    """
    data = pd.read_csv('./column_3C.dat', sep='\s+', names=model_var, header=None)
    df = data[model_var]
    return df
```

```python
from sklearn.preprocessing import LabelEncoder
#target = str(Target)
def coding_target(df,target):
    label = LabelEncoder() 
    df['Targetn'] = label.fit_transform(np.ravel(df[[target]]))
    return df
```
